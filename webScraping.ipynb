```
from bs4 import BeautifulSoup
import requests
```

```
# here i deal with local web
# now i will open the html fill with open
with open('deals.html','r') as f:
  # now i will read the content and the read method get the hole content of the page 
  content = f.read()
  # now we will use the beautiful soup
  
  soup = BeautifulSoup(content,'lxml')
  # now i will get all the card inside the page using it's class
  cards_single_items = soup.find_all('div',class_='card single-item')
  for card in cards_single_items:
    # now i will get all the places with it's price
    place_name = card.find_all('div',class_='store-item-name')
    place_price = card.span.text
    for i in place_name:
      print(f'{i.text} coast{place_price}')
```

```
def find_job():

  # now i will deal with real web using request method
  # her i will get the data from web call time job to give me all the jobs a few days ago
  html_page = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=macine+learning&txtLocation=').text

  soup = BeautifulSoup(html_page,'lxml')

  list_of_jobs = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')

  for job in list_of_jobs:

    pubilshed_date = soup.find('span',class_='sim-posted').span.text

    # here i will get all the job that posted today

    if pubilshed_date ==  'Posted today':

      job_name = job.h2.a.text

      company_names = job.find('h3',class_='joblist-comp-name').text.replace(' ','')

      skills = job.find('span',class_='srp-skills').text.replace(' ','')
      # now i want to get the link that give me more info about the job

      more_info = job.header.h2.a['href']
    
      print(f'job Name: {job_name.strip()}')

      print(f'company name: {company_names.strip()[:-10]}')

      print(f'skills: {skills.strip()}')

      print(f'more info : {more_info}')
      
      print(' ')
```

```
find_job()
```
